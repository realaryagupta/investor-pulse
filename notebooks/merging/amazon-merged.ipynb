```python
import pandas as pd
```

```python
amazon = pd.read_csv('../../data/scraped-data/AMZN_daily_2025.csv')
amazon.drop(amazon.index[0], inplace=True)
```

```python
amazon['Date'] = pd.to_datetime(amazon['Date'])
amazon['Open'] = amazon['Open'].astype(float)
amazon['High'] = amazon['High'].astype(float)
amazon['Low'] = amazon['Low'].astype(float)
amazon['Close'] = amazon['Close'].astype(float)
amazon['Volume'] = amazon['Volume'].astype(float)
```

```python
amazon_headlines = pd.read_csv('../../data/clean-data/amazon-final-cleaned-headlines.csv')
```

```python
amazon_headlines['date'] = pd.to_datetime(amazon_headlines['date'])
```

```python
reddit = pd.read_csv('../../data/clean-data/cleaned_reddit_v2.csv')
reddit.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
reddit = reddit[reddit['tickers_mentioned'].str.contains('AMZN', na=False)]
```

```python
reddit
```

```python
reddit['date'] = pd.to_datetime(reddit['date'])
```

```python
reddit['date'] = reddit['date'] + pd.Timedelta(days=1)
amazon_headlines['date'] = amazon_headlines['date'] + pd.Timedelta(days=1)
```

```python
merged = pd.merge(
    amazon,
    reddit,
    how='left',
    left_on='Date',
    right_on='date',
    suffixes=('', '_reddit')
)
merged.drop(columns=['date'], inplace=True)
```

```python
merged = pd.merge(
    merged,
    amazon_headlines,
    how='left',
    left_on='Date',
    right_on='date',
    suffixes=('', '_news')
)
merged.drop(columns=['date'], inplace=True)
```

```python
merged.sort_values('Date', inplace=True)
merged.reset_index(drop=True, inplace=True)
```

```python
merged.sample(5)
```

```python
merged['tickers_mentioned'] = 'AMZN'
```

```python
merged.drop('company', axis = 1, inplace = True)
```

```python
merged.to_csv('../../data/merged-data/amazon-merged.csv', index=False)
```
