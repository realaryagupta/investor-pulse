```python
import pandas as pd
```

```python
google = pd.read_csv('../../data/scraped-data/GOOGL_daily_2025.csv')
google.drop(google.index[0], inplace=True)
```

```python
google.head()
```

```python
google['Date'] = pd.to_datetime(google['Date'])
google['Open'] = google['Open'].astype(float)
google['High'] = google['High'].astype(float)
google['Low'] = google['Low'].astype(float)
google['Close'] = google['Close'].astype(float)
google['Volume'] = google['Volume'].astype(float)
```

```python
google_headlines = pd.read_csv('../../data/clean-data/google-final-cleaned-headlines.csv')
```

```python
google_headlines.sample(5)
```

```python
google_headlines['date'] = pd.to_datetime(google_headlines['date'])
```

```python
reddit = pd.read_csv('../../data/clean-data/cleaned_reddit_v2.csv')
reddit.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
reddit
```

```python
reddit = reddit[reddit['tickers_mentioned'].str.contains('GOOGL', na=False)]
```

```python
reddit
```

```python
reddit['date'] = pd.to_datetime(reddit['date'])
```

```python
google.info()
```

```python
google_headlines.info()
```

```python
reddit.info()
```

```python
reddit['date'] = reddit['date'] + pd.Timedelta(days=1)
google_headlines['date'] = google_headlines['date'] + pd.Timedelta(days=1)
```

```python
merged = pd.merge(
    google,
    reddit,
    how='left',
    left_on='Date',
    right_on='date',
    suffixes=('', '_reddit')
)
merged.drop(columns=['date'], inplace=True)
```

```python
merged = pd.merge(
    merged,
    google_headlines,
    how='left',
    left_on='Date',
    right_on='date',
    suffixes=('', '_news')
)
merged.drop(columns=['date'], inplace=True)
```

```python
merged.sort_values('Date', inplace=True)
merged.reset_index(drop=True, inplace=True)
```

```python
merged.sample(5)
```

```python
merged['tickers_mentioned'] = 'GOOGL'
```

```python
merged.drop('company', axis = 1, inplace = True)
```

```python
merged.to_csv('../../data/merged-data/google-merged.csv', index=False)
```
