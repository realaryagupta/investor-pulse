```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
response =  requests.get("https://news.google.com/search?q=apple%20stock%20when:7d&hl=en-IN&gl=IN&ceid=IN:en")
```

```python
response.status_code
```

```python
soup = bs(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
elements = soup.find_all('div', class_ = 'UW0SDc')
```

```python
for el in elements:
    paragraphs = el.find(class_ = 'MCAGUe')
    print(paragraphs)
```

```python
for el in elements:
    # Get the main container
    container = el.find(class_='MCAGUe')
    if container:
        # Example: Extract the author/source name
        author = container.find('div', class_='vr1PYe')
        if author:
            print("Author/Source:", author.text.strip())

        # Example: Extract the headline (often inside a <div> or <a> with a unique class)
        # You need to inspect the HTML to find the right class for the headline
        headline = container.find('button', attrs={'aria-label': True})
        if headline:
            print("Headline:", headline['aria-label'])

        # Example: Extract the date (you need to find the class or tag for date)
        # This is a placeholder; replace 'date-class' with the actual class name
        date = container.find('div', class_='date-class')
        if date:
            print("Date:", date.text.strip())
```

```python
# Find the container with class 'D9SJMe'
container = soup.find(class_='D9SJMe')
if container:
    # Find all elements inside the container that have a 'class' attribute
    for tag in container.find_all(class_=True):
        # tag['class'] is a list of class names for that tag
        print(tag['class'])
```
